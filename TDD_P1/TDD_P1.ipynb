```java test.class  
package com.codingdojo.lee;

import org.junit.Test;

import static org.hamcrest.core.Is.is;
import static org.junit.Assert.assertThat;

public class GameNumberTest {
    @Test
    public void show_number(){
        checkGameNumberOutput(1, "1");
        checkGameNumberOutput(2, "2");
    }

    @Test
    public void show_fizz_raw_number_is_divisible_by_3(){
        checkGameNumberOutput(3,"fizz");
    }

    @Test
    public void show_buzz_raw_number_is_divisible_by_5(){
        checkGameNumberOutput(5,"buzz");
    }

    @Test
    public void show_fizz_and_buzz_raw_number_is_divisible_by_35(){
        checkGameNumberOutput(15,"fizzbuzz");
    }

    @Test
    public void show_fizz_or_buzz_if_raw_number_is_divisible_by_3_or_5(){
        checkGameNumberOutput(13,"fizz");
    }

    private void checkGameNumberOutput(int i, String s) {
        assertThat(new GameNumber(i).toString(), is(s));
    }
}


```

```源码文件
package com.codingdojo.lee;

import static java.lang.String.valueOf;

public class GameNumber {

    private int rawNumber;

    public GameNumber(int rawNumber){
        this.rawNumber = rawNumber;
    }

    @Override
    public String toString(){

        if(isRelatedTo(3) && isRelatedTo(5)){
            return "fizzbuzz";
        }

        if(isRelatedTo(3)){
            return "fizz";
        }

        if(isRelatedTo(5)){
            return "buzz";
        }

        return valueOf(rawNumber);
    }

    private boolean isRelatedTo(int i) {
        return rawNumber %i ==0 || valueOf(rawNumber).contains(valueOf(i));
    }
}

```
